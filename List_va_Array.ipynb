```python
import matplotlib.pyplot as plt

resistance = [1, 2, 3, 4, 5]
voltage = [10, 20, 30, 40, 50]
current = [v / r for v, r in zip(voltage, resistance)]

plt.plot(resistance, current, marker='o')
plt.xlabel("Qarshilik (Ohm)")
plt.ylabel("Tok (A)")
plt.title("Ohm qonuni grafik")
plt.grid(True)
plt.show()
```

```python
V = 10  # Volt
R = 5   # Ohm
I = V / R
print(f"Tok: {I} A")
```

```python
from ipywidgets import interact

def ohms_law(voltage, resistance):
    current = voltage / resistance
    print(f"Tok: {current:.2f} A")

interact(ohms_law, voltage=(1, 100, 1), resistance=(1, 20, 1))
```

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from ipywidgets import interactive

# Magnit maydonining kuch chiziqlari uchun funksiya
def plot_magnetic_field(strength, angle):
    # Koordinatalar
    x = np.linspace(-5, 5, 20)
    y = np.linspace(-5, 5, 20)
    X, Y = np.meshgrid(x, y)

    # Magnit maydonini hisoblash
    Bx = strength * np.cos(angle) * np.ones_like(X)
    By = strength * np.sin(angle) * np.ones_like(Y)

    # Grafik
    plt.figure(figsize=(6, 6))
    plt.streamplot(X, Y, Bx, By, color='r', linewidth=2)
    plt.xlim(-5, 5)
    plt.ylim(-5, 5)
    plt.title(f'Magnetic Field - Strength: {strength}, Angle: {angle}')
    plt.xlabel('X axis')
    plt.ylabel('Y axis')
    plt.grid(True)
    plt.show()

# Interaktiv elementlar
strength_slider = widgets.FloatSlider(value=1.0, min=0.1, max=5.0, step=0.1, description='Strength:')
angle_slider = widgets.FloatSlider(value=0.0, min=0.0, max=360.0, step=1.0, description='Angle:')

# Interaktiv vizualizatsiya
interactive_plot = interactive(plot_magnetic_field, strength=strength_slider, angle=angle_slider)
display(interactive_plot)
```
